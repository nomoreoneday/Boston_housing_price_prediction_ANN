### 网络结构：

* l1 = Linear(X, W1, b1)
* s1 = Relu(l1)
* l2 = Linear(s1, W2, b2)
* cost = MSE(y, l2)
* 隐藏层维度为10

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
data = load_boston()
X_ = data['data']
y = data['target']
y = y.reshape(y.shape[0],1)
#数据规范化
X_ =(X_ - np.mean(X_,axis = 0)) / np.std(X_, axis = 0)
```

```python
n_features = X_.shape[1]
n_hidden = 10 #隐藏维度为10层
w1 = np.random.randn(n_features,n_hidden)
w2 = np.random.randn(n_hidden,1)
b1 = np.zeros(n_hidden)
b2 = np.zeros(1)

#激活函数
def Relu(x):
    result = np.where(x<0,0,x)
    return result

def MSE_loss(y,y_hat):
    return np.mean(np.square(y_hat-y))

def Linear(X,w,b):
    y = X.dot(w) + b
    return y
```

```python
learning_rate = 1e-6
for t in range(5000):
    l1 = Linear(X_,w1,b1)
    s1 = Relu(l1)
    y_pred = Linear(s1,w2,b2)
    
    #计算损失函数
    loss = MSE_loss(y,y_pred)
    #print(t,loss)
    
    #反向传播
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = s1.T.dot(grad_y_pred)
    grad_temp_relu = grad_y_pred.dot(w2.T)
    grad_temp_relu[l1 < 0] = 0
    grad_w1 = X_.T.dot(grad_temp_relu)
    
    #权重更新
    w1 = pd.DataFrame(w1 - learning_rate * grad_w1)
    w2 = pd.DataFrame(w2 - learning_rate * grad_w2)
    
```
